### Now I want to calculate the success rate by different features

```python
object_feature_list = list(bank.dtypes[bank.dtypes == 'object'].index)

for feature in object_feature_list[:-1]:
    
    feature_perct =  feature_perc(feature)
    plt.title('Success rate by {}'.format(feature))
    plot_bar(feature_perct.sort_values(ascending= False),5,10, text_color = 'blue')
    plt.show()
```

```python
categories = bank['y'].unique()
bin_range = (bank['age'].min(),bank['age'].max())

for c in categories:
    plt.hist(bank[bank['y']==c]['age'],alpha=0.4,label=c',range=bin_range,bins=30,density=True)
    
plt.legend()
plt.title('Age distribution among each target class')
plt.ylabel('counts')
plt.xlabel('age')
txt = '''Fig.1 The X axis is age and y is the count of each class in target varibles.
From the plot, we could see that the younger people and the elder people tends to subscribe the campaign'''
plt.figtext(0.5, -0.1, txt, wrap=False, horizontalalignment='center', fontsize=12)
#plt.show()
plt.savefig("figures/age_y.png",bbox_inches='tight')
plt.clf()
```

```python
#time to produce a matrix
count_matrix = bank.groupby(['education', 'y']).size().unstack()
count_matrix_norm = count_matrix.div(count_matrix.sum(axis=1),axis=0)
print(count_matrix_norm)
```

```python
count_matrix_norm.plot(kind='bar', stacked=True)

plt.ylabel('fraction of target class')
plt.legend(loc=4)
plt.title('Fraction of target class in each education condition')
txt = '''Fig.2 X is education condition and y is fraction of target class.
From the plot, it seems that people who are illiterate tent to subscribe more to the term deposit '''
#now we organize our plot and make it pretty
plt.figtext(0.5, -0.25, txt, wrap=False, horizontalalignment='center', fontsize=12)
plt.show()
plt.savefig("figures/education_y.png",bbox_inches='tight')
plt.clf()
```
